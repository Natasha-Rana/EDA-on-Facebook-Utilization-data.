<a id=section602></a> 
### 6.2. Missing Data and its imputation

- Here, we are going to **substitute/replace** the **missing values** with a **central tendancy measure (mean, mode, median).**

````python
fb_df['gender'].nunique()    # Checking for no. of unique values in gender column.
````

````python
fb_df['gender'].unique()    # Checking for unique values in gender column.
````

````python
fb_df['gender'].mode()      # Calculating mode of gender column.
````

````python
fb_df['gender'] = fb_df['gender'].fillna(fb_df['gender'].mode()[0])  # Filling missing values in gender column with its mode.
print(fb_df['gender'].unique())
````

````python
fb_df.isnull().sum()      # To confirm that there are no missing values in the gender column now.
````

````python
fb_df['tenure'].median()     # Calculating median of tenure column.
````

````python
fb_df['tenure'] = fb_df['tenure'].fillna(fb_df['tenure'].median())  # Filling missing values in tenure column with its median.
print(fb_df['tenure'].nunique())
````

````python
fb_df.isnull().sum()   # Final check to confirm that there are no missing values remaining in the dataset.
````

* Now, we can see that there are **no missing values** in the dataset.

<a id=section603></a> 
### 6.3. Treatment of Outliers

- We have **outliers in age column of the dataset.** We need to **remove/treat these outliers to get clean data.**
- **Removing/treating outliers** is **extremely important** because:
- **Outliers** can **increase variability in data**, thereby **decreasing the statsitical power of the data.**
- **Outliers** can also **decrease** the **success rate** of **machine learning models.**
- **Presence of outliers** can **lead to wrong predictions**.

````python
fb_df[fb_df['age'] > 80].count()       # Checking for values where age is greater than 80 years.
````

````python
def clean_age(age):                                 # Creating a new column "age_clean" that replaces out-of-range ages with "NaN"
    if age >= 13 and age <= 80:                    # The oldest living person on record lived to be 122 years, 164 days (Jeanne Calment, 1875-1997)
        return age
    else:
        return np.nan
fb_df['age_clean'] = fb_df['age'].apply(clean_age)

print("'age'")                                                          # Check out the new column and make sure it looks right
print("Minimum value: ", fb_df["age_clean"].min())
print("Maximum value: ", fb_df["age_clean"].max())
print("How many values are NaN?: ", pd.isnull(fb_df['age_clean']).sum())
print("Frequency table for age_clean:")
fb_df["age_clean"].value_counts().sort_index(0)
````

* Here, we have **set the maximum age limit to 80.** This is because **if we set the maximum age limit to 90, the outliers were still not completely removed.**
* Here, we have **deliberately introduced NaN/null values** so that we can later on **replace** these values with the **median of age column.**

````python
fb_df.isnull().sum()           # Checking for missing values in the dataset.
````

* Here, we can see that **null values are introduced** in the **new column 'age_clean'.**

````python
fb_df.info()                           # Getting basic info. about the dataset after the requisite changes have been made.
````

````python
sns.set(color_codes=True)                                      # Plot a histogram of the users' ages.
sns.set_palette(sns.color_palette("muted"))

sns.distplot(fb_df["age"]);
````

````python
sns.set(color_codes=True)                                # Plot a histogram of the age_clean column (remove any NaN values)
sns.set_palette(sns.color_palette("muted"))

sns.distplot(fb_df["age_clean"].dropna());
````

* Here, we have plotted a **distribution plot** to **check** the **distribution** of the **new column age_clean.** This **distribution** seems **much smoother from that of age column.** This **distribution does not contain outliers**, which were shown previously in the age column.

````python
# Plotting KDE plot to see the relationship between age and age_clean columns.
sns.pairplot(data=fb_df[['age', 'age_clean']], size=2.5, diag_kind='kde')  
````

* We are comparing both the variables: age and age_clean using pairplots. 
* We can observe that **age variable contains outliers.** Whereas, **age_clean** has a **smoother distribution curve** and **does not contain outliers.**
* Both the variables: **age and age_clean are positively correlated.**

````python
age_boxplot = sns.boxplot(x=fb_df["age"])     # Drawing boxplot for age column which contains outliers.
````

````python
age_clean_boxplot = sns.boxplot(x=fb_df["age_clean"]) # Drawing boxplot to confirm that there are no outliers in the age_clean column.
````

* From the above boxplots, we can observe that **age_clean is free from outliers but age still contains outliers.** We would **apply clean_age function** to the **age column,** to ensure that the **age column also gets free from outliers.**

````python
fb_df['age'] = fb_df['age'].apply(clean_age)     # Applying clean_age function to age column. 
````

````python
fb_df['age'].describe()     # Checking descriptive statsitical measures for age column.
````

````python
fb_df['age_clean'].describe()    # Checking descriptive statsitical measures for age_clean column.
````

* Both variables age and age_clean show same results.

* Now we are **replacing the NaN values of the age column** with its **median** because **median is less likely to get affected by the presence of outliers.**
* The **main reason for not replacing the missing values with the mean age** is that **mean gets drastically affected by the presence of outliers.**


````python
fb_df['age'] = fb_df['age'].fillna(fb_df['age'].median())  # Filling missing values in age column with its median.
print(fb_df['age'].nunique())
````

````python
fb_df.isnull().sum()            # Checking for missing values in the dataset after requiste changes have been made.
````

````python
sns.set(color_codes=True)                                                # Plot a histogram of the users' ages.
sns.set_palette(sns.color_palette("muted"))

sns.distplot(fb_df["age"]);
````

* Now **after replacing NaN values of age column with its median,** we have **successfully removed the outliers.**
* The **distribution** for the **age column** has now being **transformed** to a **bi-modal distribution**, i.e., **it has two high peaks.**
* The **reason for this transformation** is that when we had **replaced the NAN values of the age column with its median(27 years), the concentration of data points increased for this age(27 years) which gave rise to another peak in this distribution.**

````python
fb_df.head(20)     # Getting a glimpse of the dataset after requisite changes have been made.
````

````python
fb_df.info()     # Getting basic info. about the dataset after the changes.
````

* Similarly, **we would drop userid, dob_day, dob_year, dob_month columns as they will not fetch us any results.**

````python
fb_df.drop(['userid','dob_day','dob_month', 'dob_year', 'age_clean'], axis = 1, inplace = True)  # Dropping columns from the original dataset.
````
